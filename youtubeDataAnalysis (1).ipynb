```python
c=widgets.Dropdown(
    options=video['genre'].unique(),
    value='Sports',
    description='genre:',
    disabled=False,
)
def graph(a):
  dr=video[video['genre']==a]
  gt=dr.groupby('channelTitle')
  gt=gt.sum().sort_values(by='view_count',ascending=False).head()
  sns.barplot(y=gt.index,x=gt['view_count'])
  plt.title(a)
  plt.xlabel('views')
  plt.ylabel('channels')
  plt.show()
```

```python
widgets.interact(graph,a=c)
```

```python
widgets.interact(graphs,a=c)
```

```python
c=widgets.Dropdown(
    options=video['genre'].unique(),
    value='Sports',
    description='genre:',
    disabled=False,
)

def genre(z):
  gen=video[video['genre']==z]
  d=widgets.Dropdown(
  options=gen['channelTitle'].unique(),
  value=gen['channelTitle'].unique()[0],
  description='channel:',
  disabled=False,)
  def channel(b):
    gt=gen.groupby('title')
    gt=gt.max().sort_values(by='view_count',ascending=False).head()
    sns.barplot(y=gt.index,x=gt['view_count'])
    plt.title(b)
    plt.xlabel('views')
    plt.ylabel('video')
    plt.show()
  widgets.interact(channel,b=d)

```

```python
widgets.interact(genre,z=c)
```

```python
c=widgets.Dropdown(
    options=data['channelTitle'].unique(),
    value=data['channelTitle'].unique()[0],
    description='channel:',
    disabled=False,
)
def channel(z):
  bt=data[data['channelTitle']==z]
  b=widgets.Dropdown(
  options=bt['title'].unique(),
  value=bt['title'].unique()[0],
  description='video:',
  disabled=False,)
  def fun(t):
    gt=bt[bt['title']==t]
    sns.lineplot(y=gt['trending_date'],x=gt['view_count'])
    plt.title(t)
    plt.xlabel('views')
    plt.ylabel('trending')
    plt.show()
  widgets.interact(fun,t=b)

```

```python
widgets.interact(channel,z=c)
```
